```python
import cv2
from matplotlib import pyplot as plt 
import numpy as np
```

```python
image = cv2.imread("201804110000031010.tif",cv2.IMREAD_COLOR)
```

```python
image.shape
```

```python
GB = cv2.GaussianBlur(image,(11,11),0)

plt.imshow(GB)
```

```python
gray = cv2.cvtColor(GB, cv2.COLOR_BGR2GRAY)

plt.imshow(gray)
```

```python
canny = cv2.Canny(gray, 30, 150)

plt.imshow(canny)
```

```python
ret, img_binary = cv2.threshold(canny, 127, 255, 0)
plt.imshow(img_binary)
```

```python
contours, hierarchy = cv2.findContours(img_binary, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
```

```python
res_list = []
```

```python
for cnt in contours:
    #cv2.drawContours(roi, [cnt], 0, (255, 0, 0), 3)  # blue
    if cv2.contourArea(cnt) < 1500: 
        cv2.drawContours(img_binary, [cnt], -1, 255, -1)
        x,y,w,h = cv2.boundingRect(cnt)
        if w>150 and h>150:
            res_list.append(image[y: y + 150,x: x + 150])
            #plt.imshow(image[y: y + 50, x: x + 50])
```
